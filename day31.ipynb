## Email Sending

```python
import json
import smtplib
```

```python
with open('config.json') as f:
    config = json.load(f)[0]
```

```python
user = config.get('username')
pwd = config.get('password')
```

```python
user
```

```python
server = smtplib.SMTP('smtp.gmail.com',587)
server.ehlo()
server.starttls()
```

```python
to = 'prernasharma0429@gmail.com'
sub = 'Sample Email from Python'
msg = '''Hi Prerna
This is a test mail from Python

Thanks'''
```

```python
msg_body = '\r\n'.join(['To: %s'%to, 'From: %s'%user, 'Subject: %s'%sub, '', msg])
```

```python
print(msg_body)
```

```python
server.login(user, pwd)
server.sendmail(user, [to], msg_body)
print('Sent')
```

## Read Emails

```python
import imaplib
```

```python
mail = imaplib.IMAP4_SSL('smtp.gmail.com')
mail.login(user,pwd)
mail.select('inbox')
```

```python
_, emails = mail.search(None, 'All')

email_indx = emails[0].decode().split()
email_indx
```

```python
_, data = mail.fetch(email_indx[0], '(RFC822)')
print(data[0][-1].decode())
```

```python
_, data = mail.fetch(email_indx[2], '(UID BODY[TEXT])')
print(data[0][-1].decode())
```
