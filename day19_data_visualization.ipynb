```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.info()
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.countplot(x='sex', data=tips, hue='smoker')
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.mean,ci=95)
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,hue='time')
```

```python
sns.boxplot(x='smoker',y='tip',data=tips)
```

```python
plt.figure(figsize=(15,6))
sns.boxplot(x='sex',y='tip',data=tips, hue='day')
```

```python
sns.set_style('darkgrid')
```

```python
plt.figure(figsize=(15,6))
sns.violinplot(x='sex',y='tip',data=tips, hue='day')
```

```python
plt.figure(figsize=(15,6))
sns.stripplot(x='sex',y='tip',data=tips, hue='day')
```

```python
plt.figure(figsize=(15,6))
sns.swarmplot(x='sex',y='tip',data=tips, hue='day')
```

```python
plt.figure(figsize=(15,6))
sns.displot(x='tip',data=tips)
```

```python
plt.figure(figsize=(15,6))
sns.rugplot(x='tip',data=tips)
```
